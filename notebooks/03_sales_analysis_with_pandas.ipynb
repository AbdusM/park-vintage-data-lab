# Mission 3: Sales Analysis with pandas
Combine sales and inventory and answer practical questions.

```python
import pandas as pd
from pathlib import Path

repo_root = Path.cwd()
if repo_root.name == "notebooks":
    repo_root = repo_root.parent
```

```python
sales = pd.read_csv(repo_root / 'data' / 'raw' / 'sales.csv')
inventory = pd.read_csv(repo_root / 'data' / 'raw' / 'inventory.csv')
merged = sales.merge(inventory, on='item_id', how='left')
merged[['sale_id', 'product_name', 'category', 'sale_price', 'channel']].head()
```

```python
merged['revenue'] = merged['sale_price']
merged['profit'] = merged['sale_price'] - merged['cost']
merged['profit'].sum()
```

```python
by_channel = merged.groupby('channel')['revenue'].sum().sort_values(ascending=False)
by_channel
```

```python
by_category = merged.groupby('category')['revenue'].sum().sort_values(ascending=False)
by_category
```
